```python
import subprocess
from IPython.core.magic import register_cell_magic

@register_cell_magic
def java(line, cell):
    # 1. Salva o código em um arquivo .java
    # Importante: O nome do arquivo deve ser Main.java se a classe for Main
    with open('Main.java', 'w') as f:
        f.write(cell)

    # 2. Compila o código usando javac
    compile_process = subprocess.run(['javac', 'Main.java'], capture_output=True, text=True)

    if compile_process.returncode != 0:
        print("❌ Erro de Compilação:")
        print(compile_process.stderr)
        return

    # 3. Executa o bytecode usando java
    run_process = subprocess.run(['java', 'Main'], capture_output=True, text=True)
    print(run_process.stdout)
    if run_process.stderr:
        print("⚠️ Erro de Execução:", run_process.stderr)
```

```python
%%java
import java.net.InetAddress;

public class Main {
    public static void main(String[] args) {
        // No Colab, o Gateway padrão geralmente é o .1
        String gatewayIp = "172.28.0.1";
        int testes = 5; // Vamos fazer 5 testes seguidos

        System.out.println("--- 📊 MONITOR DE PERFORMANCE DA REDE ---");
        System.out.println("Alvo: Gateway Padrão (" + gatewayIp + ")");

        try {
            InetAddress roteador = InetAddress.getByName(gatewayIp);

            for (int i = 1; i <= testes; i++) {
                long tempoInicio = System.currentTimeMillis(); // Marca a hora de saída

                if (roteador.isReachable(2000)) { // Tenta alcançar o roteador
                    long tempoFim = System.currentTimeMillis(); // Marca a hora de chegada
                    long latencia = tempoFim - tempoInicio;

                    System.out.println("Teste #" + i + ": Resposta em " + latencia + "ms ✅");
                } else {
                    System.out.println("Teste #" + i + ": FALHA (Timeout) ❌");
                }

                Thread.sleep(1000); // Espera 1 segundo entre os testes
            }
        } catch (Exception e) {
            System.out.println("Erro: " + e.getMessage());
        }
        System.out.println("------------------------------------------");
    }
}
```
